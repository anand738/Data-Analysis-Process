### Pandas Export

- to csv
- to excel
- to html
- to json
- to sql

```python
import pandas as pd
df = pd.DataFrame()
```

## to_csv

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
temp_df = df.groupby('batter')['batsman_run'].sum().reset_index()
temp_df
```

```python
temp_df.to_csv('batsman_runs.csv',index=False)
```

```python
df.pivot_table(index='batter',columns='BowlingTeam',values='batsman_run',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

## to_excel

```python
df
```

```python
temp_df = df.groupby('batter')['batsman_run'].sum().reset_index()
```

```python
temp_df
```

```python
temp_df.to_excel("batsman_runs.xlsx")
```

```python
temp_df.to_excel("output.xlsx",sheet_name='batsman_run')
```

```python
temp_df2 = df.pivot_table(index='batter',columns='BowlingTeam',values='batsman_run',aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:  
    temp_df.to_excel(writer, sheet_name='Sheet_name_1')
    temp_df2.to_excel(writer, sheet_name='Sheet_name_2')
```

## to_html

```python
df
```

```python
df.query('batsman_run == 6').pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count').to_html('sixes_heatmap.html')
```

## to_json

```python
df
```

```python
df.groupby(['BattingTeam','batter'])['batsman_run'].sum().unstack().to_json('ipl.json')
```

## to_sql

```python
df
```

```python
import pymysql
from sqlalchemy import create_engine
```

```python
engine = create_engine("mysql+pymysql://root:@localhost/ipl")
# {root}:{password}@{url}/{database}
df.to_sql('ipl_delivery', con = engine, if_exists = 'append')
```

```python
temp_df.to_sql('batsman_runs', con = engine, if_exists = 'append')
```

```python
six_df = df.query('batsman_runs == 6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count')
```
